

*  To read the image we use imread() function from cv2 library. By default, the images are read in the form BGR( Blue Green Red). Therefore we used cvtColor() function to convert it into RGB form.
* imshow() function to display the image.
***INPUT IMAGE**





```python
img = cv2.imread('virat kohli.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```



* Converting the original image to grayscale using the cvtColor().



```python
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey, cmap='gray')
```

```python
smooth = cv2.medianBlur(grey, ksize=5)
plt.imshow(smooth, cmap='gray')
```

* highlighting edges to cartoonize the image.







```python
get_edge = cv2.adaptiveThreshold(smooth, maxValue=255, adaptiveMethod=cv2.ADAPTIVE_THRESH_MEAN_C, thresholdType=cv2.THRESH_BINARY, blockSize=9, C=9)
plt.imshow(get_edge, cmap='gray')
```



* applying the bilateral filter to remove the noise.
* keep edge sharp.



```python
color_image = cv2.bilateralFilter(img, d=9, sigmaColor=300, sigmaSpace=300)
plt.imshow(color_image, cmap='gray')
```

```python
cartoon_image = cv2.bitwise_and(src1=color_image, src2=color_image, mask=get_edge)
plt.imshow(cartoon_image, cmap='gray')
```
